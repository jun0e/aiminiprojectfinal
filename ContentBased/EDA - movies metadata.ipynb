```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./movies-data/combine_data.csv")
df.head()
```

```python
df.info()
```

```python
df = df.drop(["Unnamed: 0", "Unnamed: 0.1"], axis=1)
```

```python
df.describe()
```

```python
df.isna().any()
```

```python
df.isna().sum()
```

```python
numerical_data = ['belongs_to_collection','budget', 'id', 'popularity', 'revenue', 'runtime',
                    'vote_average', 'vote_count', 'profit', 'weighted_rating']
categorical_data = ['original_language', 'release_date', 'release_year', 'status', 'title']
list_data = ['genres', 'production_countries', 'production_companies']
df[numerical_data].hist(bins= 50, figsize=(15,10))
```

```python
for col in numerical_data:
    sns.distplot(df[col], kde=True)
    plt.show()
```

```python
for col in numerical_data:
    sns.boxplot(x=df[col])
    plt.show()
```

```python
for col in categorical_data:
    print('-'*50)
    print(f'The column name is: {col}')
    print(df[col].value_counts())
```

```python
for label in categorical_data:
    if(label is not 'release_date'):
       sns.barplot(x= df[label].value_counts()[:10].index,
           y= df[label].value_counts()[:10].values)
       plt.xlabel(label)
       plt.xticks(rotation=60)
       plt.ylabel("Frequency")
       plt.title('Show of {} Bar Plot'.format(label))
       plt.show()
    
```
